```python
# import dependencies 
import matplotlib.pyplot as plt 
import statistics 
```

```python
# set the x-axis to a list of strings for each month 
x_axis = ['Jan', 'Feb', 'March','April','May','June','July','Aug','Sep','Oct', 'Nov','Dec']
# set the y-axis to a list of floats as the total fare in US dollars accumulated for each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot 
plt.plot(x_axis, y_axis)
plt.show()
```

```python
 # create the plot with ax.plt() 
fig, ax = plt.subplots() 
#returns a tuple that contains a figure and axes object(s), unpacks the tuple into the variables

ax.plot(x_axis, y_axis)
```

```python
# create the plot and add a label for the legend 
plt.plot(x_axis, y_axis, marker ='d',color = 'green', linewidth = 2, label ="Boston")

#create labels for the x,y axes 

plt.xlabel("Date")
plt.ylabel("Fare($)")

#set the y limit between 0 and 45 
plt.ylim(0,45)

#add the grid line 
plt.grid()

#create the title 
plt.title("Pyber Fare by Month")

#add the legend 
plt.legend()
plt.show()
```

```python
# create the bar plot 
plt.bar(x_axis,y_axis)
```

```python
# create the bar plot 
plt.bar(x_axis, y_axis, color ='green', label ="Boston")

#create labels for the x,y axes 
plt.xlabel("Date")
plt.ylabel("Fare($)")

#create the title 
plt.title("Pyber Fare by Month")

#add the legend 
plt.legend()
plt.show()
```

```python
# horzintal bar graph 
plt.barh(x_axis, y_axis)
```

```python
#gca() -> get current axes 
plt.barh(x_axis, y_axis, color = 'magenta', label ="Boston")
plt.gca().invert_yaxis()

#create labels for the x,y axes 
plt.xlabel("Fare($)")
plt.ylabel("Date")

#create the title 
plt.title("Pyber Fare by Month")
plt.legend()

plt.show()
```

```python
# create the plot with ax.plt() 
fig, ax = plt.subplots() 
ax.bar(x_axis, y_axis,color ='k', label ='New York')
ax.legend()

plt.show()
```

```python
# create the plot with ax.plt() 
fig, ax = plt.subplots() 
ax.barh(x_axis, y_axis, color ='cyan', label ='Chicago')
ax.legend()


plt.gca().invert_yaxis()
```

```python
# create the scatter plot 
plt.scatter(y_axis, x_axis, color ='red', label ='Chicago')

#create labels for the x,y axes 
plt.xlabel("Fare($)")
plt.ylabel("Date")

plt.xlim(0,50)
plt.gca().invert_yaxis()

#create the title 
plt.title("Pyber Fare by Month")
plt.legend()

plt.show()
```

```python
y_axis_larger = []

for data in y_axis: 
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s = y_axis_larger)
```

```python
# create the plot with ax.plt() 
fig, ax = plt.subplots() 

y_axis_five = []
for data in y_axis: 
    y_axis_five.append(data*5)
    
ax.scatter(y_axis, x_axis, s= y_axis_five, color ='skyblue', label ='Boston', 
           edgecolors ='k', linewidth = 2, alpha = 0.8)

plt.legend()
plt.xlim(0,50)
plt.xlabel("Fare($)")
plt.ylabel('Date')
plt.title('PyBer Fare by Month')
plt.gca().invert_yaxis()

plt.show()
```

```python
# create the pie chart 
plt.pie(y_axis, labels =x_axis)

plt.show()
```

```python
explode = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.subplots(figsize =(8,8))

plt.pie(y_axis,labels=x_axis, explode=explode, colors = colors,  autopct='%.1f%%')
plt.show()
```

```python
# create the plot with ax.plt() 
fig, ax = plt.subplots() 
    
ax.pie(y_axis, labels = x_axis)

plt.show()
```

```python
# create the plot with ax.plt() 
fig, ax = plt.subplots(figsize = (8,8)) 
explode = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow",
          "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

ax.pie(y_axis, labels = x_axis, autopct='%.1f%%', explode = explode, 
       colors = colors, startangle = 90, shadow = True, counterclock = False)

plt.show()
```

```python
# get the standard deviation of the values in the y-axis 
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr = stdev)
```

```python
plt.errorbar(x_axis, y_axis, yerr = stdev, capsize=3)
```

```python
plt.bar(x_axis, y_axis, yerr = stdev, capsize=3)
```

```python
import numpy as np 
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0,51, step = 5.0))
plt.gca().invert_yaxis()
```

```python
from matplotlib.ticker import MultipleLocator 

# increase the size of the plot figure 
fig, ax = plt.subplots(figsize = (8,8)) 
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0,51, step = 5.0))

# create minor tickes ar an increment of 1 
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
